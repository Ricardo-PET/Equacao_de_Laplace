**Método da relaxação**

Nosso objetivo é solucionar numericamente a equação de Laplace, que tem a forma:

$$\nabla V = 0$$

Onde $V$, para esse projeto, é o potencial elétrico a ser encontrado. Considerando um sistema cartesiano em duas dimensões "x" e "y", a equação de Laplace pode ser escrita como:

$$\dfrac{\partial^2 V(x,y)}{\partial x^2} + \dfrac{\partial^2 V(x,y)}{\partial y^2} = 0$$

Nosso método de resolução vai envolver criar uma grade de pontos na forma de uma matriz 2D, em cada entrada será guardado o valor do potencial V naquele ponto de coordenadas (x,y) da matriz. Inicialmente só existirão os potenciais das condições de contorno e os outros pontos terão potencial zero, nosso programa atualizará todos os pontos do espaço até que a matriz gere uma solução para a equação de Laplace. Para resolver as derivadas do segundo grau da equação acima é preciso primeiro escolher um intervalo h entre os pontos da função, como vamos trabalhar sob uma matriz, os menores intervalos possíves vão ser os intervalos entre os pontos das matriz, vamos chamar este intervalo de "a". Assim, de acordo com o capítulo 5 do livro [1], a fórmula geral para uma derivada parcial de segunda ordem, de uma função V(x,y) com relação a "x" e a "y" são dadas por:

$$\dfrac{\partial^2 V(x,y)}{\partial x^2} = \dfrac{V(x+a,y)+V(x-a,y)-2V(x,y)}{a^2}$$

$$\dfrac{\partial^2 V(x,y)}{\partial y^2} = \dfrac{V(x,y+a)+V(x,y-a)-2V(x,y)}{a^2}$$
Substituindo as duas na equação de Laplace:

$$\dfrac{V(x+a,y)+V(x-a,y)-2V(x,y)}{a^2} + \dfrac{V(x,y+a)+V(x,y-a)-2V(x,y)}{a^2} = 0$$

$$\dfrac{V(x+a,y)+V(x-a,y)+V(x,y+a)+V(x,y-a)-4V(x,y)}{a^2} = 0$$

Ou,

$$V(x,y) = \dfrac{1}{4}(V(x+a,y)+V(x-a,y)+V(x,y+a)+V(x,y-a))$$

Isso nos dá uma relação muito interessante para encontrarmos o potencial $V(x,y)$ em qualquer ponto da matriz, basta somar o potencial do ponto acima, abaixo, o da esquerda, o da direita, e por fim dividir tudo por 4. Se tivermos uma condição de contorno estável, ou seja, que o seu potencial não mude com o tempo, então só é preciso fazer com que o algoritmo use a função acima em todos os pontos da matriz repeditas vezes (ao longo de vários ciclos), até que o resultado do potencial $V$ seja satisfatório. Para descobrir se a matriz está completa, basta achar a diferença do potencial antes e depois de cada ciclo para cada ponto, se este estiver menor que determinado parâmetro o programa é encerrado. Este método é chamado de método de Jacobi, e dependendo da quatidade de entradas da matriz e das condições de contorno, o programa pode demorar muito tempo para alcançar o potencial satisfatório requerido, além de ser um método instável, muitas vezes nem chegando a convergir para um resultado. Por isso vamos usar aqui o método da super relaxação e o de Gauss-Seidel como otimizações do método de Jacobi, estes métodos não só vão fazer o potencial convergir para um resultado como também vão fazer isso de forma rápida. 

O método da super relaxação apenas acelera a velocidade com que o potencial dos pontos tendem ao seu valor final de convergência, ou de solução. Ele usa um peso $\omega$ para dizer o quão intenso deve ser a próxima variação do potencial nos pontos da matriz. Sua formula para encontrar o potencial no ponto $V(x,y)$ é dada por:

$$V(x,y) = \dfrac{1 + \omega}{4}(V(x+a,y)+V(x-a,y)+V(x,y+a)+V(x,y-a)) - \omega V(x,y)$$

O valor de $\omega$ não pode ser qualquer um, na verdade se ele for maior que 1 o método nunca convergirá para uma solução. O valor aconselhado pelo livro para o peso é $\omega = 0.9$.

Quando usamos o método de Jacobi original, nós calculamos os potenciais $V(x,y)$ pegando apenas os potencias dos pontos ao redor durante o mesmo ciclo, entretanto, no meio do ciclo, enquanto o programa está resolvendo entrada por entrada da matriz, ele vai deixando um "rastro" de potenciais atualizados, potenciais estes que estão ainda mais perto de convergir que os outros que ainda não foram atualizados na matriz. O método de Gauss-Seidel faz uso destes potenciais já atualizados no meio do clico para encontrar potenciais $V(x,y)$ ainda mais próximos da solução, gerando assim um código mais rápido e eficiente.

**Definindo as constantes**

A matriz, ou seja, nosso universo terá lado "L", vamos considerar a tolerância para a variação mínima possível do potencial para cada ponto como sendo tol = $10^{-6}$, quando o código chegar a esta tolerância ele deverá parar de calcular e gerar o resultado da matriz. Além disso vamos escolher um delta máximo como sendo a maior variação de potencial dentre todos os pontos em um mesmo ciclo, ou seja, o ponto que tiver a maior variação de potencial no clico ditará se todo o sistema está adequado ou não como solução, se este valor cair para baixo do valor da tolerância o programa para. A voltagem máxima Vmax será uma constante usada para criar a condição de contorno do problema. 

```python
L = 200 #(centimetros)
tol = 1.0e-6
Vmax = 1.0 #(Volt)
deltamax = 10.0
```

**Barra carregada e esfera metálica descarregada**

Para este sistema, vamos considerar uma barra com um potencial Vmax na horizontal e logo sobre ela uma esféra metálica descarregada, ou seja, com potencial interno zero, além disso os limites máximos da matriz (suas paredes) também vão ter potencial zero. Para que estas condições de contorno não sejam modificadas pelos ciclos do programa, vamos definir a região com a barra e com a esfera como sendo regiões proibidas para a varredura da matriz "phi".

```python
# Definindo as condições iniciais
phi = zeros([L,L],float)
phi[int(0.6*L),:] = Vmax     #Coordenadas da barra
x0, y0, R = 0.4*L,0.5*L,0.1*L   #Coordenadas do centro da esfera e seu raio

# Atualizando o potencial (o potencial nas bordas é fixo)
while(deltamax>tol):
    delta = []
    for i in range(1,L-1):
        for j in range(1,L-1):
            if ((i-x0)**2 + (j-y0)**2)**(1/2) > R and i != 0.6*L:
                a = phi[i,j]
                phi[i,j] = (1.9/4)*(phi[i+1,j]+phi[i-1,j]+phi[i,j+1]+phi[i,j-1])-0.9*phi[i,j]
                delta.append(abs(phi[i,j]-a))
    deltamax = max(delta)        
imshow(phi)
colorbar()
show()
```
